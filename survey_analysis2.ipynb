```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', 30)
```

```python
df = pd.read_csv('cys.csv')
```

```python
df.tail()
```

```python
df.shape
```

```python
df1 = df.groupby('Level of study (Undergraduate):  Foundation Year == 0;       1st Year = = 1       2nd Years == 2;    3rd Years ==  3;     4th Years ==  4').mean()
```

```python
df2 = df1.reset_index() # reseting the index
df2
```

Changing the columns to usable rows variables

```python
tidy = pd.melt(df2, id_vars = ['Level of study (Undergraduate):  Foundation Year == 0;       1st Year = = 1       2nd Years == 2;    3rd Years ==  3;     4th Years ==  4'], var_name='question', value_name='avg_answer')
tidy
```

```python
# using seaborn as its is easier than matplotlib
plt.figure(figsize=(30, 10))
sns.barplot(x="question", hue="Level of study (Undergraduate):  Foundation Year == 0;       1st Year = = 1       2nd Years == 2;    3rd Years ==  3;     4th Years ==  4", y="avg_answer", data=tidy)
plt.show()
```

```python
# matplotlib using pivot table to make it able to make a fast plot
tidy.pivot("question", "Level of study (Undergraduate):  Foundation Year == 0;       1st Year = = 1       2nd Years == 2;    3rd Years ==  3;     4th Years ==  4", "avg_answer").plot(kind='bar')
plt.show()
```
